# Convert filtered VCF files to HDF5

This document converts `SeqArray`/`SNPRelate` filtered VCF files and converts
them to HDF5 objects for use with `Ipyrad` (specifically STRUCTURE).

```python
import ipyrad.analysis as ipa
from pathlib import Path
```

```python
dirpath = Path("../results/variant-filtering")  # Replace with your actual directory

for vcf in dirpath.glob("*.vcf.gz"):
    if vcf.is_file():  # Check if the item is a file
        basename = vcf.stem.split(".")[0]
        dirname = vcf.parent

        # Unfiltered Ipyrad output
        converter = ipa.vcf_to_hdf5(
            name=f"{basename}",
            data=vcf.as_posix(),
            workdir=dirname,
            ld_block_size=50000
        )

        # run the converter
        converter.run(force=True)
```
